## 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
## Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
## You need to find following details:
## A) Rank
## B) Name
## C) Artist
## D) Upload date
## E) Views


```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(6)
```

```python
Rank = []

for i in range(0,2):
    rk = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]".format(i))

for r in rk:

    Rank.append(r.text)
Rank
```

```python
video_name = []

for v in range(0,2):
    vn = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]".format(v))

for v in vn:

    video_name.append(v.text)
video_name
```

```python
artist = []

for a in range(0,2):
    ar = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]".format(a))

for a in ar:

    artist.append(a.text)
artist
```

```python
Date = []

for d in range(0,2):
    de = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]".format(d))

for d in de:

    Date.append(d.text)
Date
```

```python
views = []

for v in range(0,2):
    vs = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]".format(v))

for v in vs:

    views.append(v.text)
views
```

```python
import pandas as pd
youtube=pd.DataFrame({})
youtube['Rank']=Rank
youtube['Video Name']=video_name
youtube['Artist ']=artist
youtube['Views']=views
youtube['Date']=Date
youtube
```
